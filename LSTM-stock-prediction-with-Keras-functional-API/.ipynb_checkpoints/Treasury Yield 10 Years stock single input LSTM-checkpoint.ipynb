```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dataset_train = pd.read_csv('TNX.csv')
dataset_train
```

```python
dataset_train = dataset_train.dropna()
dataset_train
```

```python
training_set = dataset_train.iloc[:, 0:2]
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
tr=training_set.iloc[:, 1].values.reshape(-1, 1)
training_set.iloc[:, 1] = sc.fit_transform(tr)
```

```python
training_set.plot()
```

```python
training_set
```

```python
for s in range(1, 15):
    training_set['shift_{}'.format(s)] = training_set['Open'].shift(s)
    training_set['shift_{}'.format(s)] = training_set['Open'].shift(s)
```

```python
training_set.head(15)
```

```python
X_train = training_set.dropna().drop('Open', axis=1)
y_train = training_set.dropna()[['Open']]
```

```python
X_train
```

```python
X_trin=X_train.iloc[:180,:].drop('Date', axis=1)
X_test=X_train.iloc[181:,:].drop('Date', axis=1)
y_trin=y_train.iloc[:180,:]
y_test=y_train.iloc[181:,:]
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping
```

```python
K.clear_session()

model = Sequential()
model.add(Dense(12, input_dim=14, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)
```

```python
model.fit(X_trin, y_trin, epochs=200,
          batch_size=1, verbose=1, callbacks=[early_stop])
```

```python
y_pred = model.predict(X_test)
plt.plot(y_test.values)
plt.plot(y_pred)
```

```python
X_train = X_trin.values
X_test= X_test.values

y_train = y_trin.values
y_test = y_test.values
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 1, 14)
X_test_t = X_test.reshape(X_test.shape[0], 1, 14)
```

```python

from keras.layers import LSTM
```

```python
K.clear_session()
model = Sequential()

model.add(LSTM(6, input_shape=(1, 14)))

model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train_t, y_train,
          epochs=100, batch_size=1, verbose=1,
          callbacks=[early_stop])
```

```python
y_pred = model.predict(X_test_t)
plt.plot(y_test)
plt.plot(y_pred)
```

```python
from keras.models import Input, Model
from keras.layers import Dense
inputs = Input(shape=(1, 14))
x = LSTM(6)(inputs)
output = Dense(1, activation = 'sigmoid')(x)
n_net = Model(inputs, output )
n_net.compile(loss='mean_squared_error', optimizer='adam')

n_net.fit(x=X_train_t, y=y_train, epochs=10, batch_size=1, verbose=1,
          callbacks=[early_stop])
```
