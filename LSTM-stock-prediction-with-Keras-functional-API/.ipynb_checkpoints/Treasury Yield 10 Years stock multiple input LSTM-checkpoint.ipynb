```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data1 = pd.read_csv('EURUSDX.csv')
data2= pd.read_csv('TNX.csv')
data1= data1.dropna()
data2= data2.dropna()
```

```python
data1 = data1.iloc[:, 0:2]
data2 = data2.iloc[:, 0:2]
```

```python
data1=data1.merge(data2,on='Date').drop('Open_y', axis=1)
```

```python
data2
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
data1.iloc[:, 1] = sc.fit_transform(data1.iloc[:, 1].values.reshape(-1, 1))
data2.iloc[:, 1] = sc.fit_transform(data2.iloc[:, 1].values.reshape(-1, 1))
```

```python
data1.plot()
```

```python
data2.plot()
```

```python
for s in range(1, 15):
    data1['shift_{}'.format(s)] = data1['Open_x'].shift(s)
    data2['shift_{}'.format(s)] = data2['Open'].shift(s)
```

```python
X_data1 = data1.dropna().drop(['Open_x',"Date"], axis=1)
y_data1 = data1.dropna()[['Open_x']]
```

```python
X_data2 = data2.dropna().drop(['Open',"Date"], axis=1)
y_data2 = data2.dropna()[['Open']]
```

```python
X_data1_train=X_data1.iloc[:200,:]
X_data1_test =X_data1.iloc[201:,:]
y_data1_train=y_data1.iloc[:200,:]
y_data1_test =y_data1.iloc[201:,:]
```

```python
X_data2_train=X_data2.iloc[:200,:]
X_data2_test= X_data2.iloc[202:,:]
y_data2_train=y_data2.iloc[:200,:]
y_data2_test= y_data2.iloc[201:,:]
```

```python
X_data2_test.shape
```

```python
X_data1_train= X_data1_train.values
X_data1_test = X_data1_test.values
y_data1_train= y_data1_train.values
y_data1_test = y_data1_test.values
```

```python
X_data2_train= X_data2_train.values
X_data2_test = X_data2_test.values
y_data2_train= y_data2_train.values
y_data2_test = y_data2_test.values
```

```python
X_data2_train.shape
```

```python
X_data1_train_t = X_data1_train.reshape(X_data1_train.shape[0], 1, 14)
X_data1_test_t  = X_data1_test.reshape (X_data1_test.shape[0], 1, 14)
X_data2_train_t = X_data2_train.reshape(X_data2_train.shape[0], 1, 14)
X_data2_test_t  = X_data2_test.reshape (X_data2_test.shape[0], 1, 14)
```

```python
X_data2_train_t.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping
from keras.layers import LSTM
from keras.models import Input, Model
from keras.layers import Dense
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)
```

```python

inputs = Input(shape=(1, 14))
x = LSTM(6)(inputs)
output = Dense(1, activation = 'sigmoid')(x)
n_net = Model(inputs, output )
n_net.compile(loss='mean_squared_error', optimizer='adam')

n_net.fit(x=X_data1_train_t, y=y_data1_train, epochs=10, batch_size=1, verbose=1,
          callbacks=[early_stop])
```

```python
from keras.layers import concatenate
from keras.layers import Dropout

input1 = Input(shape=(1,14)) # for the three columns of dat_train
x1 = LSTM(6)(input1)

input2 = Input(shape=(1,14))
x2 = LSTM(6)(input2)

con = concatenate(inputs = [x1,x2] ) # merge in metadata
x3 = Dense(50)(con)
x3 = Dropout(0.3)(x3)
output = Dense(1, activation='sigmoid')(x3)
n_net = Model(inputs=[input1, input2], outputs=output)
n_net.compile(loss='mean_squared_error', optimizer='adam')
```

```python
n_net.fit(x=[X_data2_train_t, X_data1_train_t], y=y_data1_train, epochs=10, batch_size=1, verbose=1,
          callbacks=[early_stop])
```

```python
y_pred = n_net.predict([X_data2_test_t,X_data1_test_t])
plt.plot(y_data1_test)
plt.plot(y_pred)
```
